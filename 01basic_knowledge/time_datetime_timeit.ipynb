## time时间的三种类型，
1. 时间戳是计算机能够识别的时间；
2. 时间字符串是人能够看懂的时间；
3. 元组,结构时间是用来操作时间的


## 时间戳-->结构化时间
1. time.gmtime(时间戳)    #UTC时间，与英国伦敦当地时间一致
2. time.localtime(时间戳) #当地时间。例如我们现在在北京执行这个方法：与UTC时间相差8小时，UTC时间+8小时 = 北京时间 

```python
aa1 = time.gmtime(1500000000)
aa2 = time.localtime(1500000000)
print("aa1",aa1)
print("aa2",aa2)
```

## 结构化时间-->时间戳

```python
struct_time1 = time.localtime(1500000002) # 使用时间戳，生成一个结构化时间
timestamp1 = time.mktime(struct_time1)
print("struct_time1",struct_time1)
print("timestamp1",timestamp1)
```

## 结构化时间-->转字符串时间
1. 结构化时间 --> %a %b %d %H:%M:%S %Y串
2. time.asctime(结构化时间) 如果不传参数，直接返回当前时间的格式化

```python
struct_time2 = time.localtime()
str_time1 = time.strftime("%Y-%m-%d %X")
str_time2 = time.strftime("%Y-%m-%d %X",struct_time2)
print("str_time1",str_time1)
print("str_time2", str_time2)
```

```python
# 返回当前时间的字符串
asctime_b = time.asctime()
# 使用struct_time时间进行转化
print(time.asctime(struct_time2))
print(asctime_b)
print(type(asctime_b))
```

## 计算时间差

```python
import time
time_last=time.mktime(time.strptime('2008-08-08 20:30:00','%Y-%m-%d %H:%M:%S'))
time_now=time.mktime(time.strptime('2019-10-01 06:00:00','%Y-%m-%d %H:%M:%S'))
dif_time=time_now-time_last
time_struct=time.gmtime(dif_time)
print(time_struct)
print('2008年8月8日20点30分到2019年10月1日06点过去了%s年%s月%s天%s小时%s分%s秒'%(time_struct.tm_year-1970,time_struct.tm_mon-1,time_struct.tm_mday-1,time_struct.tm_hour,time_struct.tm_min,time_struct.tm_sec))
```

## timedelta

```python
from datetime import timedelta
```

timedelta 类型可进行加减乘除，以及大小判断

timedelta 对象表示两个 date 或者 time 的时间间隔
1. datetime.timedelta(days=0, seconds=0, microseconds=0, milliseconds=0, minutes=0, hours=0, weeks=0)
2. 所有参数都是可选的并且默认为 0。 这些参数可以是整数或者浮点数，也可以是正数或者负数。
3. 只有 days, seconds 和 microseconds 会存储在内部。 参数单位的换算规则如下：
    * 1毫秒会转换成1000微秒。
    * 1分钟会转换成60秒。
    * 1小时会转换成3600秒。
    * 1星期会转换成7天。

```python
delta = timedelta(days=10, seconds=1, microseconds=0, milliseconds=0, minutes=0, hours=2, weeks=10)
```

```python
print(delta)
print(delta.days)
print(delta.seconds)
print(delta.microseconds)
```

### date.weekday()
返回一个整数代表星期几，星期一为0，星期天为6。例如， date(2002, 12, 4).weekday() == 2，表示的是星期三。

### date.isoweekday()
返回一个整数代表星期几，星期一为1，星期天为7

## date.isocalendar()¶
返回一个三元元组，(ISO year, ISO week number, ISO weekday) 。

## date.isoformat()¶
返回一个以 ISO 8601 格式 YYYY-MM-DD 来表示日期的字符串:

```python
date1 = date.today()
print(date1.weekday())
print(date1.isoweekday())
print(date1.isocalendar())
print(date1.isoformat())
```

## 计算距离特定事件的天数

```python
import time
import datetime
today = date.today()
my_birthday = date(today.year,1,1)
if my_birthday < today:
    my_birthday = my_birthday.replace(year=today.year + 1)
# abs时间取正数
time_to_birthday = abs(my_birthday - today)
print(time_to_birthday.days)
```

```python
from datetime import date
d = date.fromordinal(730920) # 730920th day after 1. 1. 0001
print(d.isoformat())
print(d.strftime("%d/%m/%y"))
print(d.strftime("%A %d. %B %Y"))
print(d.ctime())
t = d.timetuple()
print(t)
print(d.isocalendar())
print(d.replace(year=2005))
```
